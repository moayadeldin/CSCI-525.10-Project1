```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/NYC-Airbnb-2023.csv")
```

```python
df[df['price']>1000]
```

```python
filtered_df = df[df["minimum_nights"] >= 2]

print(len(filtered_df["minimum_nights"]))
print(max(filtered_df["minimum_nights"]))

plt.figure(figsize=(10, 6))
plt.hist(filtered_df["minimum_nights"], bins=200, edgecolor="black")
plt.xlabel("Minimum Nights")
plt.ylabel("Frequency")
plt.title("Minimum Nights Distribution (Ignoring Values repeated only once)")
plt.savefig("fig.png")
plt.show()
```
